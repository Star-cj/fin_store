```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel(r'Final_Output.xlsx')
df.head(2)
```

                    COMPARING DIFFERENT OPTION BASED ON GENDER

```python
# the whole female in different position 
female=df[(df['Gender']=="Female")&(df['Position'])]
female=female['Position'].value_counts()
# female
```

```python
# Total number of male in different postion of the company 
male=df[(df['Gender']=="Male")&(df['Position'])]
male=male['Position'].value_counts()
# male
```

```python
#Comparing the male and female workers based on position

plt.plot(male, label='Male',linestyle='--')     # '--' for long broken line, ':' for dotted lines, '-.' for broken and dotted lines 
plt.plot(female, label='Female',linestyle='-.')
plt.title("Compare gender base on position", size=20)
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(5,3))
x=np.arange(male.count())
width=0.4
ax=fig.add_axes([0,0,1,1])

ax.bar(x+0.0,male, label='Male',color='r',width=width)
ax.bar(x+0.4,female, label='Female',color='g',width=width)

plt.title("Compare gender base on Primary Division", size=20)
plt.xlabel("Position",color='black')
plt.legend()
plt.show()
```

```python
# the whole female in different Primary Division 
female_pri=df[(df['Gender']=="Female")&(df['Division Primary'])]
female_pri=female_pri['Division Primary'].value_counts()

# the whole male in different Primary Division 
male_pri=df[(df['Gender']=="Male")&(df['Division Primary'])]
male_pri=male_pri['Division Primary'].value_counts()
# print(male_pri)
# female_pri

fig,(ax1,ax2)=plt.subplots(1,2, figsize=(25,8))
fig.suptitle('Dotted for Male: Orange line for Female', size=20)
ax1.plot(male_pri,'o')  #you can use 'o','+' or even 'tab:green'
ax2.plot(female_pri,'tab:orange')
plt.xlabel("Primary Divisions",color='r',size=15)
```

```python
# the whole female with different respondents
female_resp=df[(df['Gender']=="Female")&(df['Respondents'])]
female_resp=female_resp['Respondents'].value_counts().sort_index(ascending=False)

# the whole male with different respondents
male_resp=df[(df['Gender']=="Male")&(df['Respondents'])]
male_resp=male_resp['Respondents'].value_counts().sort_index(ascending=False)
# print(male_resp)
# print(female_resp)

plt.plot(male_resp, label='Male',linestyle='-.',color='red')  
plt.plot(female_resp, label='Female',linestyle=':',color='black')
plt.title("Compare gender base on Respondents", size=20)
plt.xlabel("Respondents",color='green')
plt.legend()
plt.show()
```

```python
female_resp=df[(df['Gender']=="Female")&(df['Same Answer'])]
female_resp=female_resp['Same Answer'].value_counts().sort_index(ascending=False)

male_resp=df[(df['Gender']=="Male")&(df['Same Answer'])]
male_resp=male_resp['Same Answer'].value_counts().sort_index(ascending=False)
# print(male_resp)
# print(female_resp)

plt.plot(male_resp, label='Male',linestyle='-.',color='blue')  
plt.plot(female_resp, label='Female',linestyle='--',color='red')
plt.title("Compare gender base on Answer Gotten", size=20)
plt.xlabel("Answer Selected",color='purple')
plt.legend()
plt.show()
```

```python
female_resp=df[(df['Gender']=="Female")&(df['Employment Type'])]
female_resp=female_resp['Employment Type'].value_counts()

male_resp=df[(df['Gender']=="Male")&(df['Employment Type'])]
male_resp=male_resp['Employment Type'].value_counts()
# print(male_resp)
# print(female_resp)

fig=plt.figure(figsize=(4,4))
x=np.arange(male_resp.count())
width=0.25
ax=fig.add_axes([0,0,1,1])

ax.bar(x+0.0,male_resp, label='Male',color='r',width=width)
ax.bar(x+0.25,female_resp, label='Female',color='g',width=width)

plt.title("Compare gender base on Employment Type", size=20)
plt.legend()
plt.show()
```
